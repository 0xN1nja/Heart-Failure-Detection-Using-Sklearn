```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

# Create Dataset

```python
df=pd.read_csv(r"C:\Users\speed\Documents\Heart Failure Detection Using Sklearn\data\heart_failure_clinical_records_dataset.csv")
```

```python
x=df.drop("DEATH_EVENT",axis=1)
```

```python
y=df["DEATH_EVENT"]
```

# Create And Train Model

```python
model=linear_model.LogisticRegression()
```

```python
model.fit(x,y)
```

# Predict And Write To Dataframe Object

```python
df["predicted"]=model.predict(df.drop("DEATH_EVENT",axis=1))
```
